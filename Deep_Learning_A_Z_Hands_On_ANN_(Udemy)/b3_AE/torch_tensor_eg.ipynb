```python
ratings = pd.read_csv('ml-1m/ratings.dat', sep = '::', header = None,
                      engine = 'python', encoding = 'latin-1')


print('ratings.shape = {}'.format(ratings.shape))
ratings.head()
```

```python
# Preparing the training set and the test set
training_set = pd.read_csv('ml-100k/u1.base', delimiter = '\t',header=None)
training_set = np.array(training_set, dtype = 'int')
test_set = pd.read_csv('ml-100k/u1.test', delimiter = '\t',header=None)
test_set = np.array(test_set, dtype = 'int')

pd.DataFrame(training_set).head(10)
```

```python
# Getting the number of users and movies
nb_users = int(max(max(training_set[:,0]), max(test_set[:,0])))
nb_movies = int(max(max(training_set[:,1]), max(test_set[:,1])))

print('nb_users = {} and nb_movies = {}'.format(nb_users, nb_movies))
```

```python
# Converting the data into an array with users in lines and movies in columns
# in the given data user id starts from 1 not from 0.
#
def convert(data):
    new_data = []
    for id_users in range(1, nb_users + 1):
        id_movies = data[:,1][data[:,0] == id_users]
        id_ratings = data[:,2][data[:,0] == id_users]
        ratings = np.zeros(nb_movies)
        ratings[id_movies - 1] = id_ratings
        new_data.append(list(ratings))
    return new_data
training_set = convert(training_set) # list of list, there are nb_users lists.
test_set = convert(test_set)

print(np.array(training_set).shape)
pd.DataFrame(training_set).head()

# there is list for each nb_users
# first user is 1
# from above, user 1 rated the movie 1=5, 2=3, 3=4 and he may not have rated
# some movies which will be -1.
#
# in the table below we can see the same thing.
```

```python
# Preparing the training set and the test set
tr = pd.read_csv('ml-100k/u1.base', delimiter = '\t',header=None,dtype=int,
                names=['user','movie','rating','id'])
ts = pd.read_csv('ml-100k/u1.test', delimiter = '\t',header=None,dtype=int,
                names=['user','movie','rating','id'])


print('tr.shape = {}, \nts.shape = {}'.format(tr.shape,ts.shape))
display(tr.head(10), ts.head(10))
```

```python
n_users = int(max(tr.iloc[:,0].max(), ts.iloc[:,0].max()))
n_movies = int(max(tr.iloc[:,1].max(), tr.iloc[:,1].max()))
               
print('n_users = {} and n_movies = {}'.format(n_users,n_movies))
```

```python
# get one row for user1 with all the movie ratings
```

```python
def ratings_4users(df, n_users, n_movies):
    movie = np.array(df.groupby('user')['movie'].apply(lambda x: x.tolist()).tolist())
    rating = np.array(df.groupby('user')['rating'].apply(lambda x: x.tolist()).tolist())
    rating_lst = np.zeros((n_users,n_movies), dtype=float)

    for i,u in enumerate(df['user'].unique()):
        rating_lst[u-1][ np.array(movie[i])-1] = rating[i]

    return rating_lst

tr2 = ratings_4users(tr, n_users, n_movies)
tr2
```
