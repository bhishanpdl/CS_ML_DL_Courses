```python
# mse
se1 =  (0.9-1)**2 + (0.1 - 0)**2 
se2 =  (0.1-0)**2 + (0.9 - 1)**2
se3 =  (0.4-1)**2 + (0.6 - 0)**2
se = se1 + se2 + se3

mse = 1/3 * se
mse
```

```python
# cross entropies
from numpy import log

hpq1 = - (log(0.9) * 1 + log(0.1) * 0)
hpq2 = - (log(0.1) * 0 + log(0.9) * 1)
hpq3 = - (log(0.4) * 1 + log(0.6) * 0)

Hpq = (hpq1 + hpq2 + hpq3) /3
Hpq
```
