**Question**  

	Test for two means. Independent samples, population variance known									
	New Cleaning Detergent									
										
	Background: Ajax releases a new dish detergent, called Extreme Ajax+, which they advertise to be able to clean many more dishes with a single bottle.									
	You want to see if that is true and have bought 25 bottles of Extreme Ajax+ and 25 bottles of Ajax and collected data on how many dirty dishes you can wash with a single bottle. (Population variance is assumed to be known.)									
										
	Task 1	Find the standard deviations of the two samples								
	Task 2	State the Null Hypothesis								
	Task 3	Determine the Z-score								
	Task 4	Find the p-value								
	Task 5	Interpret the result								
										
										
		Extreme Ajax+	Ajax	Difference						
	Size	25	25	?						
	Mean	115	100	15						
	Variance	625	400	6.40						
										
	Standard deviation									
	Hypothesized diff.									
	Z-score									
	p-value									

```python
n1,n2 = 25, 25
xbar1, xbar2 = 115, 100
var1, var2 = 625, 400
std1, std2 = np.sqrt(var1), np.sqrt(var2)

std1, std2
```

```python
std_err = np.sqrt(var1/n1 + var2/n2)
std_err
```

```python
# Question: Extreme Ajax+ cleans many more dishes than Ajax
# H0: Extreme cleans less than Regular Ajax
#
# Ans: 
# H0: mu(Extreme) - mu(regular) <= 0
```

```python
# since sample is large we use z-dist
Z = (xbar1-xbar2) / std_err
Z
```

```python
# p-value
nsided = 1
p_value = scipy.stats.norm.sf(np.abs(Z))*nsided # sf is survival function
round(p_value,2)
```

```python
# Result
# for 1% confidence level, 0.009 < 0.01, reject H0, accept the claim.
# also, accept the claim for 5%, and 10%.
```
