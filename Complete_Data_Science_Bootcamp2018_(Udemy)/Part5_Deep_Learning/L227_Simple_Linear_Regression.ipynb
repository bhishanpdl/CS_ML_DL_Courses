# Imports

```python
import numpy as np
 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
from mpl_toolkits.mplot3d import Axes3D
```

# Data

```python
n = 10000

xs = np.random.uniform(-10, 10, size=(n,1))
zs = np.random.uniform(-10, 10, (n,1))

X = np.c_[xs,zs]

noise = np.random.uniform(-1, 1, (n,1))
y = 2*xs - 3*zs + 5 + noise

X.shape, y.shape
```

# plot X

```python
# reshape for plotting
y = y.reshape(n,)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, zs, y)

# labels
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')

ax.view_init(azim=100)
plt.show()

# reshape back
y = y.reshape(n,1)
```

# Weights

```python
init_range = 0.1

W = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))
b = np.random.uniform(low=-init_range, high=init_range, size=1)

# learning rate
eta = 0.01

W,b
```

# Gradient descent

```python
losses = []
for i in range (100):
    
    yhat = np.dot(X,W) + b
    deltas =  yhat - y
        
    loss = np.sum(deltas ** 2) / 2 / n
    losses.append(loss)
    
    deltas_scaled = deltas / n
    
    W = W - eta * np.dot(X.T,deltas_scaled)
    b = b - eta * np.sum(deltas_scaled)
    
```
