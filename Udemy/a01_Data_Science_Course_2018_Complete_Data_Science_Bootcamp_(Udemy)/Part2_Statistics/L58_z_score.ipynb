```python
df = pd.read_excel(data_dir + '3.9.Population-variance-known-z-score-exercise.xlsx',
                  sheet_name = 'CI',
                  skiprows = 9,
                  usecols = "B")


print(df.head(5))
print()
print(df.tail(5))
print(df.shape)
```

```python
# task 1: mean and standard error
df.mean().values, df.std().values, sigma/np.sqrt(df.size)
```

```python
# task 2: z-score for 90% confidence iterval
```

```python
z_score = scipy.stats.norm.ppf(1 - (1-0.90)/2.0)
z_score
```

```python
# task 3: find confidence interval
```

```python
sigma = 15000 # given
n = df.size
xbar = df.mean()
z_score = scipy.stats.norm.ppf(1 - (1-0.90)/2.0)
std_error = sigma/np.sqrt(n)

ci_low = xbar - z_score * std_error
ci_high = xbar + z_score * std_error

ci_low, ci_high
```
