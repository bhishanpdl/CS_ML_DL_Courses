### Sampling distribution

$$
\text{sample normal distribution} N \sim \left( \mu , \frac { \sigma ^ { 2 } } { n } \right)
$$

$\bar{x}$ is the point estimate of the population mean $\mu$.\\
$s$ is the point estimate of population variance $\sigma$.

standard error is given by $\frac{\sigma}{\sqrt{n}}$.

confidence interval is given by
$$
[ \bar{x}- z_{\alpha/2} \frac{\sigma}{\sqrt{n}}, \bar{x} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}} ]
$$

For confidence level 95% alpha is 5%.  
z_alpha is found by standard normal distribution table.

Example:

confidence interval = 95%  
alpha = 0.05  
1 - alpha/2 = 1 - 0.025 = 0.975  
z_0.025 = 1.9 + 0.006 = 1.96  (1.9 and 0.06 are found from z-table)  

commonly used term for Z is critical value.
